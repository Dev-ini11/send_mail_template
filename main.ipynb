```python
def read_file(file_path: str):
    with open(file_path, "r", encoding = "utf-8") as f:
        output = f.read()
    return output
```

```python
read_file("data/Birthday_Wish_Template.txt")
```

```python
def modify(output: str, name: str, your_name: str):
    new_output = output.replace("{name}", name).replace("[Your Name]", your_name)
    
    return new_output
```

```python
print(modify(read_file("data/Birthday_Wish_Template.txt"), "Taofeek", "Inioluwa"))
```
